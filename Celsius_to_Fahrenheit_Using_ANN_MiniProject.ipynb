#PROBLEM STATEMENT
In this project, we will build a simple machine learning model to convert from
celsius to fahrenheit.

The equation is as follows T(f)° = T(℃)X9/5+32

For example, lets convert 0°C celsius temperature to Fahrenheit: (0℃X9/5)+32 = 32℉.






#**STEP** #0 : **IMPORT** **LIBRARIES**

```python
import tensorflow as tf
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#STEP #1 : IMPORT DATASETS

```python
from google.colab import files
Data = files.upload()
```

```python
df = pd.read_csv('Celsius+to+Fahrenheit.csv')
df.head()
```

```python
df.tail(10)
```

#STEP #2: DISPLAY THE DATA
---



```python
sns.scatterplot(data=df,x=df['Celsius'], y=df['Fahrenheit'])
```

#STEP #3 : CREATE TESTING AND TRAINING DATASET

```python
X_train = df['Celsius']
Y_train = df['Fahrenheit']
```

#STEP #4: BUILD AND TRAIN THE MODEL

```python
X_train.shape
```

```python
Y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```python
epoch_hist=model.fit(X_train, Y_train, epochs=500)
```

#STEP #4 : EVALUATE THE MODEL

```python
epoch_loss = epoch_hist.history.keys()
epoch_loss
```

```python
plt.plot(epoch_hist.history['loss'])
plt.title('Model Loss Progress During Training')

plt.xlabel('Epochs')
plt.ylabel('loss')
```

```python
model.get_weights()
```
